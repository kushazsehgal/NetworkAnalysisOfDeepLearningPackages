```python
import pandas as pd
```

```python
filtered_nn_df = pd.read_csv('../exports/filtered_nn.csv')
nn_edges_not_nn_df = pd.read_csv('../exports/nn_edges_not_nn.csv')
nn_edges_with_nn = pd.read_csv('../exports/nn_edges_with_nn.csv')
```

```python
libio_df = pd.read_pickle('../output/libio_depen_added.pkl')
libio_df['Package Name'] = [name.lower() for name in libio_df['Package Name'].to_list()]
```

```python
filtered_nn_by_betweeness = filtered_nn_df.sort_values(by = 'betweenesscentrality',ascending=False)['package name'][:10].to_list()
filtered_nn_by_degree = filtered_nn_df.sort_values(by='Degree',ascending=False)['package name'][:10].to_list()
nn_edges_not_nn_by_betweeness = nn_edges_not_nn_df.sort_values(by = 'betweenesscentrality',ascending=False)['package name'][:10].to_list()
nn_edges_not_nn_by_degree = nn_edges_not_nn_df.sort_values(by='Degree',ascending=False)['package name'][:10].to_list()
nn_edges_with_nn_by_betweeness = nn_edges_with_nn.sort_values(by = 'betweenesscentrality',ascending=False)['package name'][:10].to_list()
nn_edges_with_nn_by_degree = nn_edges_with_nn.sort_values(by='Degree',ascending=False)['package name'][:10].to_list()
# filtered_nn_by_betweeness = filtered_nn_df.sort_values(by = 'betweenesscentrality',ascending=False)['package name'][:10].to_list()
```

```python
for pkg_name in filtered_nn_by_betweeness:
    df = libio_df[libio_df['Package Name'].isin(filtered_nn_by_betweeness)]
    df.to_excel('../exports/filtered_nn_by_betweeness.xlsx',index = False)
for pkg_name in filtered_nn_by_degree:
    df = libio_df[libio_df['Package Name'].isin(filtered_nn_by_degree)]
    df.to_excel('../exports/filtered_nn_by_degree.xlsx',index = False)
for pkg_name in nn_edges_not_nn_by_betweeness:
    df = libio_df[libio_df['Package Name'].isin(nn_edges_not_nn_by_betweeness)]
    df.to_excel('../exports/nn_edges_not_nn_by_betweeness.xlsx',index = False)
for pkg_name in nn_edges_not_nn_by_degree:
    df = libio_df[libio_df['Package Name'].isin(nn_edges_not_nn_by_degree)]
    df.to_excel('../exports/nn_edges_not_nn_by_degree.xlsx',index = False)
for pkg_name in nn_edges_with_nn_by_betweeness:
    df = libio_df[libio_df['Package Name'].isin(nn_edges_with_nn_by_betweeness)]
    df.to_excel('../exports/nn_edges_with_nn_by_betweeness.xlsx',index = False)
for pkg_name in nn_edges_with_nn_by_degree:
    df = libio_df[libio_df['Package Name'].isin(nn_edges_with_nn_by_degree)]
    df.to_excel('../exports/nn_edges_with_nn_by_degree.xlsx',index = False)
```

```python
modularity_class = 788
cluster_names = nn_edges_with_nn[nn_edges_with_nn['modularity_class'] == modularity_class].sort_values(by = 'betweenesscentrality',ascending=False)['package name'][:10]
for pkg_name in cluster_names:
    df = libio_df[libio_df['Package Name'].isin(cluster_names)]
    df.to_excel('../exports/nn_edges_with_nn_torch_by_betweenness.xlsx',index = False)
```
